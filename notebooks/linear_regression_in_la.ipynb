```python
from numpy import array
import numpy.linalg as linalg
import matplotlib.pyplot as plt
```

```python
data = array([
    [0.05, 0.12],
    [0.18, 0.22],
    [0.31, 0.35],
    [0.42, 0.38],
    [0.5, 0.49],
])
print (data)
```

```python
X, y = data[:,0], data[:,1]
X = X.reshape((len(X), 1))
plt.scatter(X, y)
```

## Solve directly

```python
b = linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
b
```

```python
yhat = X.dot(b)
plt.scatter(X, y)
plt.plot(X, yhat, color='r')
```

## Solve using QR decomposition

```python
Q, R = linalg.qr(X)
b = linalg.inv(R).dot(Q.T).dot(y)
print (b)

# Get predictions
yhat = X.dot(b)

# Plot result
plt.scatter(X, y)
plt.plot(X, yhat, color='r')
```

## Solve with SVD
Not all equations can be solved with QR decomposition.  SVD is more stable.

$A=U\Sigma V^*$

where $A$ is an $m \times n$ matrix, $U$ is an $m \times m$ matrix, $\Sigma$ is an $m \times n$ diagonal matrix, and $V^*$ is the conjugate transpose of an $n \times n$ matrix.

The coeficients $b$ are calculated as follows

$b=X^+ . y$

where $X^+$ is the pseudoinverse of $X$, which is given by

$X^+ = U . D^+ . V^T$

where $D^+$ is the pseudoinverse of $\Sigma$ and $V^T$ is the transpose of $V^*$.

numpy's linalg.pinv returns $X^+$. 

```python
b = linalg.pinv(X).dot(y)
yhat = X.dot(b)

plt.scatter(X, y)
plt.plot(X, yhat, color='r')
```
